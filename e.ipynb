```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parmetros del ejemplo
n = 1000
xi = np.random.normal(size=n)
b_n = np.arange(1, n + 1)

# Clculos segn el teorema
spectral_density_0 = 1 / (2 * np.pi)  # Densidad espectral para N(0, 1)
eta = np.random.rand()  # Variable aleatoria no negativa (puedes ajustar esto)

# Convergencia en L1
conditional_expectation = np.zeros(n)
for k in range(n):
    conditional_expectation[k] = np.mean(xi[:k+1])**2

conditional_expectation /= b_n
conditional_expectation *= spectral_density_0

# Varianza normalizada de Sn
variance_sn = np.cumsum(xi)**2 / b_n**2
variance_sn *= 2 * np.pi * spectral_density_0

# Convergencia en distribucin
sn_distribution = np.cumsum(xi) / b_n
sn_distribution = (sn_distribution - np.mean(sn_distribution)) / np.std(sn_distribution)

# Grficos
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(xi)
plt.title('Secuencia Estacionaria (Xi)')

plt.subplot(2, 2, 2)
plt.plot(conditional_expectation)
plt.title('Convergencia en L1')

plt.subplot(2, 2, 3)
plt.plot(variance_sn)
plt.title('Varianza Normalizada de Sn')

plt.subplot(2, 2, 4)
plt.hist(sn_distribution, bins=30, density=True, alpha=0.7, color='blue')
x = np.linspace(min(sn_distribution), max(sn_distribution), 100)
plt.plot(x, norm.pdf(x), 'r--', label='N(0, 1)')
plt.title('Convergencia en Distribucin')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulacin de datos de ventas
np.random.seed(42)
n_obs = 100
t = np.arange(n_obs)
stationary_component = 2 * np.sin(2 * np.pi * t / 12)
noise = np.random.normal(scale=1, size=n_obs)
sales = stationary_component + noise

# Graficar los datos de ventas simulados
plt.figure(figsize=(10, 6))
plt.plot(t, sales, label='Datos de Ventas Simulados')
plt.title('Simulacin de Datos de Ventas con Componente Estacionario')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.legend()
plt.show()

# Aplicar el teorema (implementacin simplificada)
cumulative_sales = np.cumsum(sales)
normalized_cumulative_sales = cumulative_sales / np.sqrt(t + 1)

# Graficar la suma acumulativa normalizada
plt.figure(figsize=(10, 6))
plt.plot(t, normalized_cumulative_sales, label='Suma Acumulativa Normalizada')
plt.title('Suma Acumulativa Normalizada de Datos de Ventas')
plt.xlabel('Mes')
plt.ylabel('Suma Acumulativa Normalizada')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def spectral_density(x):
    # Funcin de densidad espectral (puedes definir la tuya segn tus necesidades)
    return np.exp(-x**2 / 2)

def central_limit_theorem(xi, b_n):
    n = len(xi)
    gamma = lambda j: np.sum(np.abs(np.mean(xi * np.roll(xi, shift=-j)[::-1])))

    Gamma = np.array([gamma(j) for j in range(n)])
    condition = np.sum(Gamma[:n//2]) / n

    if condition < np.inf:
        spectral_density_0 = spectral_density(0)

        eta = np.random.rand()  # Puedes ajustar esto segn tus necesidades
        result = np.sum(xi[:n])**2 / n
        result /= np.mean(b_n)
        result /= spectral_density_0

        var_sn = np.var(np.cumsum(xi))
        var_sn /= np.mean(b_n)**2
        var_sn *= 2 * np.pi * spectral_density_0

        sn_dist = np.cumsum(xi) / np.mean(b_n)
        sn_dist = (sn_dist - np.mean(sn_dist)) / np.std(sn_dist)

        return result, var_sn, sn_dist, eta * norm.rvs()

# Simulacin de datos de ventas mensuales ficticios
np.random.seed(42)
ventas_mensuales = np.random.normal(loc=100, scale=20, size=120)

# Aplicar el teorema a los datos simulados
b_n = np.arange(1, len(ventas_mensuales) + 1)
result, var_sn, sn_dist, sample = central_limit_theorem(ventas_mensuales, b_n)

# Mostrar resultados
print(f"Result: {result}, Var(S_n): {var_sn}")

# Graficar la suma acumulativa normalizada y la muestra de la distribucin
plt.figure(figsize=(12, 6))
plt.plot(sn_dist, label='Suma Acumulativa Normalizada')
plt.axhline(y=sample, color='r', linestyle='--', label='Muestra N(0, 1)')
plt.title('Aplicacin del Teorema del Lmite Central a Datos de Ventas Simulados')
plt.xlabel('Mes')
plt.ylabel('Valor')
plt.legend()
plt.show()
```
